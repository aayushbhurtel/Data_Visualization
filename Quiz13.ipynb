And for education, yes we can assign an order on basis of primary, secondary and tertiary. 

```python
from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories=['Primary', 'Secondary', "Tertiary"], ordered=True)

dfClass = df.education.astype(cat_type)

dfClass.dtype
```
