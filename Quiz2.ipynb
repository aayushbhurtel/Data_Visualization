#5. Melt the data keeping two columns of values and the type of flower

pandas library provides us with melt function which can be used to categorize and display small portion of large dataset with filters. 

Here, id_vars is also called as identifier variables which remains entact or does not change. value_vars is the list of columns to be displayed in a melted dataset. In this case, we have two columns(Sepallength and SepalWidth) to be displayed so we pass them as a parameter to value_vars. Finally, var_name is the name of a new variable column which holds the reshaped data values.

```python
df_melt = dataFrame.melt(id_vars="Class", value_vars=['SepalLength', 'SepalWidth'], var_name='columns')
df_melt
```

#6. Select flowers that have pedal length greater than 5 and sepal length less than 7

I used a loc method in pandas library which access group of values from the dataset based on the values provided. Here, I selected to display a set of data from a dataset which has value more then 5 on Sepallength column. Also, I used and operator to display another set of data whose value is less then 7, listed under PedalLength column.


```python
dataFrame.loc[(dataFrame['PedalLength'] > 5) & dataFrame['SepalLength'] < 7]
```

#7. Group the data by type of flower and compute mean and median

Here, I used groupby method or function in pandas library which groups dataFrame data using series of columns. This function groups large amount of data and computes various operation like mean and median which we do in this block of code.

I have passed a label as a 'Class' so that pandas will group my dataset based on the values in that label. Now, I will use a agg function using dot operator and pass mean and median as a parameter to calulate mean and median of those class values respectively.

```python
dataFrame.groupby(by ='Class').agg(['mean'],['median'])
```
