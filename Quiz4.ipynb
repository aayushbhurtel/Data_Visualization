# Compute the average time for a trip based on starting point

I am going to calcualte average time for a trip based on starting point which can be achieved by calculating average of duration in minutes using AVG function and group those data using GROUP BY function in bigQuery. Below is the demonstration.

```python
%%bigquery --project data-viz-big-query
SELECT start_station_name, AVG(duration_minutes) as average_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```

#Compute how many trips start at each starting point.

Again, we have to compute how many trips starts at each starting point. For this, we need to count the number of trips that starts at each of the start station. we can simply select start station name and do a COUNT on how many trips start over there.

```python
%%bigquery --project data-viz-big-query
SELECT start_station_name, COUNT(*) as no_of_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```
