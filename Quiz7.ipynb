I am loading austin_crime dataset from google public queries dataset. Displaying 10 rows of data to get an idea of how the data is arranged and can be manupulated in the dataset.

```python
%%bigquery --project data-viz-big-query
SELECT *
FROM `bigquery-public-data.austin_crime.crime`
LIMIT 5
```

These data looks interesting but I don't need all of them. I am going to only list the type of crime that occured in austin and total number of those crimes.

```python
%%bigquery --project data-viz-big-query
SELECT COUNT(*) as total_cases, description as crime_type
FROM `bigquery-public-data.austin_crime.crime`
GROUP BY description
ORDER BY total_cases DESC
LIMIT 10
```

I decided to compare Austin crime with chicago crime so here is a database for chicago crime.

```python
%%bigquery --project data-viz-big-query
SELECT *
FROM `bigquery-public-data.chicago_crime.crime`
LIMIT 5
```

Here is the formatted crime numbers and types on CHICAGO area

```python
%%bigquery --project data-viz-big-query
SELECT COUNT(*) as total_cases, primary_type as crime_type
FROM `bigquery-public-data.chicago_crime.crime`
GROUP By primary_type
ORDER By total_cases DESC
LIMIT 10
```

Now I want to join two tables and create a side by side comparision.

```python
%%bigquery --project data-viz-big-query
SELECT COUNT(Austin.description) as austin_crime_numbers, Austin.description, COUNT(Chicago.primary_type) as chicago_crime_numbers, 
FROM `bigquery-public-data.chicago_crime.crime` as Chicago FULL JOIN `bigquery-public-data.austin_crime.crime` as Austin ON Chicago.primary_type = Austin.description
GROUP BY Austin.description
LIMIT 10
```
