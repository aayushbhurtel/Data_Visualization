firstly, I load all the libraries.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

I am using Marvel universe wikipedia page to get the table with all the movies in different Phases.

[https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films](https://)

```python
raw = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_parse = raw.text
soup = BeautifulSoup(html_parse,'lxml')
```

find the length of the table from a webpage

```python
len(soup.find_all('table'))
```

lets grab a table for assignment from a webpage with a class.

```python
marvelTable = soup.find_all('table', class_ = 'wikitable plainrowheaders')
# using pandas to 
dataFrame = pd.read_html(str(marvelTable))[0]
dataFrame
```

removed unncessary rows as heading from the table using droplevel method in pandas.

```python
dataFrame.columns = dataFrame.columns.droplevel(1)
dataFrame
```

Here, I am removing last column that has "NaN" as a value which does not makes sense.

```python
del dataFrame['Unnamed: 5_level_0']
dataFrame
```

Now, I'm removing rows 6 and 13 because instead of movies, it has phase information which is irrelevant. I'm using drop method from pandas which takes the index of the rows and removes them.

```python
dataFrame = dataFrame.drop(dataFrame.index[[6,13]])
dataFrame
```

since I deleted some rows, them indexes are not in right order so I am going to reindex them all. I am going to use reindex_index method of dataFrame object.

```python
dataFrame = dataFrame.reset_index(drop=True, inplace=False)
dataFrame
```
