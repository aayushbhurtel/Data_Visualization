```python
%%bigquery --project data-viz-big-query
SELECT city as city, SUM(sale_dollars) as total_sales, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2019-01-01' AND '2019-12-31' AND city = "Des Moines"
GROUP BY city, date
ORDER By date
```

```python
%%bigquery --project data-viz-big-query
SELECT city as city, SUM(sale_dollars) as total_sales, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2020-01-01' AND '2020-12-31' AND city = "Des Moines"
GROUP BY city, date
ORDER By date
```

```python
%%bigquery --project data-viz-big-query
SELECT city as city, county as county, item_description as item, SUM(sale_dollars) as total_sales, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2021-07-01' AND '2021-08-01'
GROUP BY city, county, item, date
ORDER BY date
LIMIT 10
```
